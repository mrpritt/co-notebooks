# O drinque ideal

Seja $x_i$ a quantidade do ingrediente $i$, em litros. Além disso vamos usar uma variável auxiliar $x=\sum_{i\in[n]} x_i$ que representa a quantidade total de ingredientes.

```julia
using JuMP
using GLPK
```

Nos vamos definir, arbitrariamente, $n=10$ ingredientes e gerar dados aleatórios.

```julia
n = 10
ac = 10*rand(n) # acidez
dc = 10*rand(n) # doçura
al = 10*rand(n) # álcool
r  = 100*rand(n) # custo
P  = rand(Bool,n);
```

## Modelo

```julia
m = Model(GLPK.Optimizer);
```

```julia
@variable(m, x[1:n] >= 0)
@variable(m, X >= 0);
```

```julia
@objective(m, Min, sum(r[i]*x[i] for i=1:n))
```

```julia
@constraints(m, begin
    X == sum(x[i] for i=1:n)
    X == 1
    0.2 *X <= sum(ac[i]*x[i] for i=1:n)
              sum(ac[i]*x[i] for i=1:n) <= 0.3 *X
    0.3 *X <= sum(ac[i]*x[i] for i=1:n)
              sum(ac[i]*x[i] for i=1:n) <= 0.4 *X
    0.03*X <= sum(al[i]*x[i] for i=1:n)
              sum(al[i]*x[i] for i=1:n) <= 0.04*X
    sum(x[i] for i=1:n if P[i]) >= 2*sum(x[i] for i=1:5 if !P[i])
    end
    )
```
