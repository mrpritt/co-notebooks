# Polícia de Cidade Limpa

```julia
using JuMP
using GLPK
```

Primeiramente vamos definir $p_t$, o período antes do período $t$ ($p_t=t-1$, para $t>1$, e $p_1=6$) e as necessidades mínimas

```julia
pe = [ i-1 for i=1:6]; pe[1]=6;
nm = [ 22, 55, 88, 110, 44, 33];
```

## Modelo

```julia
m = Model(GLPK.Optimizer);
```

```julia
@variable(m, x[1:6] >= 0)
```

```julia
@objective(m, Min, sum(x[i] for i=1:6))
```

```julia
@constraint(m, [i=1:6], x[i]+x[pe[i]] >= nm[i])
```
