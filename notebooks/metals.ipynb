# Companhia Siderúrigica Jericó

```julia
using JuMP
using GLPK
```

```julia
m = Model(GLPK.Optimizer);
@variable(m, x[1:2] >= 0)
```

Queremos minimizar o custo total

```julia
@objective(m, Min, 3*x[1]+6*x[2])
```

sem produzir mais que $18$ horas

```julia
@constraint(m, 3x[1]+2x[2] <= 18)
```

produzindo pelo menos $5$ quilos

```julia
@constraint(m, x[1]+x[2] >=5)
```

respeitando a relação entre sucata para aço puro

```julia
@constraint(m, x[2] <= 7.0/8*x[1])
```

e sem usar demais aço ou sucata

```julia
@constraints(m ,begin
    x[1] <= 4
    x[2] <= 7
    end)
```
